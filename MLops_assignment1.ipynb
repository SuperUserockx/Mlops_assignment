Q1. Print multiplication table from 1 to 10

```python
#write here and run the cell

# Define the size of the multiplication table
size = 10

# Print the table header
print("   ", end="")
for i in range(1, size + 1):
    print(f"{i:4}", end="")
print()

# Print the table rows
for i in range(1, size + 1):
    print(f"{i:2} ", end="")
    for j in range(1, size + 1):
        print(f"{i * j:4}", end="")
    print()
```

Write a program which write all content of a given file (test.txt) into a new file by skipping line number 5 (line5)

```python
#write here and run the cell

# Function to create a file with the provided data
def create_file():
    content = """line1
                line2
                line3
                line4
                line5
                line6
                line7
                """
    
    with open("test.txt", 'w') as file:
            file.write(content)
    print("File created")

# Function to carry out the desired operation

def read_skip():

    # Reading the original file

    with open("test.txt", 'r') as file:
        lines = file.readlines()
    print("Input file data -> ")
    for i in lines:
        print(i.strip())
    
    print("Writing files to new file")

    # Creating an output file handler
    out_file = open("new_file.txt", 'w')
    for i in range(len(lines)):

        # Skipping the 5th line
        if i != 4:
             out_file.write(lines[i])
    out_file.close()

    # Reading the new file 
    with open("new_file.txt", 'r') as file:
        lines = file.readlines()
    print("Output file data -> ")
    for i in lines:
        print(i.strip())

create_file()
read_skip()
```

Q8. Write a program to identify words that contain both letters and digits from a given input string.

    Given str1 = "Harry25 is Gryffindor Wizard50 and cursed"

    Output:
    Harry25
    Wizard50

```python
#write here and run the cell

import re

def detect_letters_digits(input_string):
    # Define a regex pattern to match words containing both letters and digits
    pattern = re.compile(r'(?:\d+[a-zA-Z]+|[a-zA-Z]+\d+)')
    
    # Find all matches in the input string
    matches = pattern.findall(input_string)
    
    # Print each matched word
    for match in matches:
        print(match)

# Given input string
str1 = "Harry25 is Gryffindor Wizard50 and cursed"

# Find and print words containing both letters and digits
print("Output:")
detect_letters_digits(str1)
```

Create a Bus object that will inherit all of the variables and methods of the parent Vehicle class and display it.

```python
#write here and run the cell
class Bus(Vehicle):
    def __init__(self, max_speed, mileage, capacity):
        # Initialize the parent class
        super().__init__(max_speed, mileage)
        # Add additional attribute specific to Bus
        self.capacity = capacity

    def print_all(self):
        super().print_all()
        print("Capacity: " + str(self.capacity))

my_bus = Bus(max_speed=80, mileage=50000, capacity=50)
my_bus.print_all()
```

Q14. Create a Bus class that inherits from the Vehicle class. Give the capacity argument of Bus.seating_capacity() a default value of 50.

    class Vehicle:

      def seating_capacity(self, capacity):

Expected Output:

```python
#write here and run the cell
class Vehicle:
    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage

    def seating_capacity(self, capacity):
        return f"The seating capacity of a {self.name} is {capacity} passengers"
    
class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        return super().seating_capacity(capacity)
    
my_bus = Bus(name="bus", max_speed=80, mileage=50000)
my_bus.seating_capacity()
```

Q15. Create a Bus subclass that inherits from the Vehicle class. By default, the fare for any vehicle is calculated as the seating capacity multiplied by 100. For instances of the Bus class, an additional 10% maintenance fee should be applied to the total fare. Therefore, the final fare for a bus instance will be the sum of the total fare plus an additional 10% of that total fare.

Note: The bus has a seating capacity of 50, so the final fare amount should be 5500. You need to override the fare() method from the Vehicle class in the Bus class.

Use the provided code for the parent Vehicle class. You should access the parent class within a method of the child class.

    class Vehicle:
      def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

      def fare(self):
        return self.capacity * 100

    class Bus(Vehicle):
        pass
        School_bus = Bus("School Volvo", 12, 50)
        print("Total Bus fare is:", School_bus.fare())

    Expected Output:

    Total Bus fare is: 5500.0

```python
#write here and run the cell

class Vehicle:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    def fare(self):
        # Calculate the base fare using the parent class method
        base_fare = super().fare()
        # Calculate the total fare including the 10% maintenance fee
        total_fare = base_fare + (0.10 * base_fare)
        return total_fare


# Create an instance of Bus
School_bus = Bus("School Volvo", 12, 50)

# Print the total fare
print("Total Bus fare is:", School_bus.fare())
```

Q17. Sort JSON keys in and write them into a file

    sampleJson = {"id" : 1, "name" : "value2", "age" : 29}

    {
          "age": 29,
          "id": 1,
          "name": "value2"
    }

```python
#write here and run the cell

import json

# Given dictionary
sampleJson = {"id": 1, "name": "value2", "age": 29}

# Convert dictionary to JSON format with sorted keys
json_data = json.dumps(sampleJson, sort_keys=True, indent=4)

# Specify the filename
filename = 'sorted_data.json'

# Write the JSON data to a file
with open(filename, 'w') as file:
    file.write(json_data)

print(f"JSON data has been written to {filename}")
```
